# Time ReSampling

```python
min_all=tesla_data.resample(rule='A').min()
```

```python
max_all=tesla_data.resample(rule='A').max()
```

```python
differerence=max_all-min_all
```

```python
differerence.plot()
```

```python
max_all['Open'].plot()
```

```python
max_all['Close'].plot()
```

```python
(max_all['Open']-min_all['Open']).plot()
```

```python
(max_all['Open']-min_all['Close']).plot()
```

```python
# quartely start frequency
tesla_data.resample(rule='QS').max()
```

```python
tesla_data.resample(rule='QS').max()['High'].plot()
```

```python
# Business End Freq
tesla_data.resample(rule='BA').max()['High'].plot()
```

```python
# Business Quarter
tesla_data.resample(rule='BQS').max().plot()
```

```python
tesla_data['Open'].resample(rule='QS').max().plot(kind='bar')
```

```python
tesla_data.resample(rule='QS').max()['Open'].plot(kind='bar')
```
